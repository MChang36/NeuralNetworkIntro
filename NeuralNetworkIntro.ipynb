```python
# set pandas dataset
import pandas as pd
pd.set_option('display.max_rows', 7)
pd.set_option('display.max_columns', 6)

#import numpy
import numpy as np

# initialize lists
data = [[2190, 2133, 6892, 18.2785, 18.1903],
        [2207, 2133, 7068, 22.3517, 22.2032],
        [1910, 2134, 6216, 17.1261, 16.5916], 
        [1987, 2134, 6154, 22.0054, 21.6215], 
        [1834, 2119, 6582, 24.9748, 24.6152], 
        [2147, 2111, 7182, 27.2596, 27.0889], 
        [2268, 2015, 6901, 30.5700, 30.3031]] 

df = pd.DataFrame(data, columns = ['Gross Profit', 'Basic Shares Outstanding', 'Revenue','high','low'])
df.index = ['2019-09-30','2019-12-31','2020-03-31','2020-06-30','2020-09-30','2020-12-31','2021-03-31']
df
```

```python
#change to percentages
pct_df = pd.DataFrame(index = df.index)
for column in df.columns:
    pct_df[column] = df[column].pct_change().mul(100).round(2)

#add Decisions
decisions = [np.nan]
for row in range(1, len(pct_df.index)):
    decisions.append(1 if int(pct_df.iloc[row,3]) >= 3 and int(pct_df.iloc[row,4]) >= 3 else 0)
pct_df['Decision'] = decisions

pct_df = pct_df.drop(labels='2019-09-30',axis=0)
pct_df
```

```python
X = pct_df.T.drop(labels=['high','low','Decision'],axis=0)
Y = pct_df[['Decision']].T

X
```

```python
# What value to initialize w, b, and alpha?
w, b, alpha = 1, 1, 1
m = len(X)

# How many times to iterate?
for count in range(10):
    Z = np.dot(w, X) + b
    print("Z: {}".format(Z))
    A = 1/(1 + np.exp(-Z))
    print("A: {}".format(A))
    dZ = A - Y.to_numpy()
    print("dz: {}".format(dZ))
    dw = (1/m*X*dZ).to_numpy()
    print("dw: {}".format(dw))
    db = 1/m * np.sum(dZ)
    print("db: {}".format(db))
    w = w - alpha*dw
    print("w: {}".format(w))
    b = b - alpha*db
    print("b: {}".format(b))
    
print(w)
print(b)
```
